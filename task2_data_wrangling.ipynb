```python
"""
Script/Notebook: data_wrangling.ipynb
Description: Performs common data wrangling operations :
             - Trimming whitespace (from column names and string values)
             - Converting columns to appropriate data types (e.g., numeric)
             - Handling missing or invalid entries
             - dropping rows with problematic data
Usage: colab

Author: Meenu Mary John
"""
```

```python
# Import necessary libraries
import pandas as pd
```

```python
def wrangle_data(file_path:str) -> None:
    """
    Clean the data in a messy CSV file.

    Args:
        file_path (str): The path to the CSV file.
    """
    try:
        df = pd.read_csv(file_path)
        print(df)
        df.columns = df.columns.str.strip()
        string_cols = df.select_dtypes(include='object').columns
        df[string_cols] = df[string_cols].apply(lambda col: col.str.strip())
        df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
        df.dropna(inplace=True)
        print(df.head())
    except Exception as e:
          print(f"Error: {e}")

wrangle_data('/content/sample_data/stage0_foundations/messy_data.csv')
```
