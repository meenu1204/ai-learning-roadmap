```python
"""
Script/Notebook: average_score.ipynb
Description: Reads a CSV file with 'name' and 'score' columns and prints the average score.
Usage: colab
"""
```

```python
# Import necessary libraries
import pandas as pd
```

```python
def calculate_average(file_path: str) -> None:
    """
    Calculate the average score from a CSV file.

    Args:
        file_path (str): The path to the CSV file.
    """
    try:
        df = pd.read_csv(file_path)
        print(df)
        if 'score' not in df.columns:
          raise ValueError("CSV must contain a 'score' column.")
        average_score = df['score'].mean()
        print(f"The average score is: {average_score}")
    except Exception as e:
          print(f"Error: {e}")

calculate_average('/content/sample_data/stage_0/scores.csv')
```
